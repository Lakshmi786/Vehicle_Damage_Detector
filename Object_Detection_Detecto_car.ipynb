preprocessing of data

```python
from torchvision import transforms
from detecto import core

transform_img = transforms.Compose([transforms.ToPILImage(),
                                   transforms.Resize(400),
                                   transforms.RandomHorizontalFlip(0.5),
                                   transforms.ToTensor(),
                                   utils.normalize_transform(),])
```

```python
dataset = detecto.core.Dataset('train.csv','train/',transform = transform_img)
```

```python
from detecto import visualize
for i in range(0,5):
  try:
    image,information = dataset[i]
  except ValueError :
    continue
  detecto.visualize.show_labeled_image(image, information['boxes'], information['labels'])
```

```python
dataloader = detecto.core.DataLoader(dataset)
dataloader
```

```python
validation_data = detecto.core.Dataset('val.csv','valid/')
```

```python
categories = ['damage','BG']
classifier = core.Model(categories)
```

```python
history = classifier.fit(dataloader,validation_data,epochs = 40,verbose =  True)
```

```python
plt.plot(history)
```
